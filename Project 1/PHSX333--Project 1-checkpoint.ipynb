```python
#--------------------------------------------------------------------------------------------------------------------------
# Part 1: Proving the steady state solutions analytically 
#--------------------------------------------------------------------------------------------------------------------------
# Goal:
#------
# The steady state solution of this system can be found by setting the left-hand side equal to zero. Show that the steady 
# state values for (X,Y ) are (A,B/A).
#
# Solution:
#----------
# Solving for the steady state solution we set
#                                
#                            dX/dt = A - (B+1)X + (X^2)Y
#                            dY/dt = BX - (X^2)Y
#
# equal to zero and solve for X and Y.
#
# For dX/dt
#
#                            A - (B+1)X + (X^2)Y = 0   
#                            *If you put in 0 for X to solve for the steady state solution you get A 
#                            as everything else goes to zero:
#                            X = A
#
# For dY/dt 
#                            BX - (X^2)Y = 0
#                            *If you put A in for X and solve for Y tou get the following for the steady state solution:
#                            AB - AAY = 0
#                            AAY = AB
#                            Y = B/A
#
#---------------------------------------------------------------------------------------------------------------------------
```

```python
#---------------------------------------------------------------------------------------------------------------------------
# Part 2: Solving the differential equations
#---------------------------------------------------------------------------------------------------------------------------
# Goal:
# -----
# Write a program to solve numerically the rate equations given above. Use a 4th-order Runge-Kutta algorithm. Your program
# should input the initial values of X and Y and the xed concentrations A and B, and print out t, X, and Y to a le
#---------------------------------------------------------------------------------------------------------------------------
# Solution:
# --------------------------------------------------------------------------------------------------------------------------

import numpy as np
from matplotlib import pylab as py

# Function taking in X,Y,A,B and the period so we can insert all componenets of the function and perfrom the integration to
# analyze the results.
def oscillationmain(X,Y,A,B,T):
    
    def f(r,t):
        X = r[0]
        Y = r[1]
        fx = A - (B+1)*X + X*X*Y
        fy = B*X - X*X*Y
        return np.array([fx,fy], float)

    # Array from start time 0 to end time of 30 with 1000 steps in between
    a = 0
    b = T
    N = 10000
    h = b/N

    tpoints = np.arange(a,b,h)
    xpoints = []
    ypoints = []
    r = np.array([X,Y], float)
    
    # 4th-Order Runge-Kutta
    for t in tpoints:
        xpoints.append(r[0])
        ypoints.append(r[1])
        k1 = h*f(r,t)
        k2 = h*f(r+0.5*k1, t+0.5*h)
        k3 = h*f(r+0.5*k2, t+0.5*h)
        k4 = h*f(r+k3, t+h)
        r += (k1+ 2*k2 + 2*k3 + k4)/6
    
    # Returns the three data arrays with calculated values
    return np.array([xpoints, ypoints, tpoints], float)
#---------------------------------------------------------------------------------------------------------------------------

#---------------------------------------------------------------------------------------------------------------------------
# Part 3: Iterations with A and B constant
#---------------------------------------------------------------------------------------------------------------------------
# Goal:
#------
# Systematically vary the initial values of X and Y for given values of A and B. Are the steady state behaviors of X and Y 
#independent of their initial conditions? Make a plot of Y versus X for one set of values for A and B but with many dierent
#initial conditions. Do this for at least three dierent values of A and B.
#---------------------------------------------------------------------------------------------------------------------------
#Solution
#---------------------------------------------------------------------------------------------------------------------------

# For loop to run though different iterations of X and Y and plot the results
        
file = open("COdata.txt","w")              # Creating a text file
for m in np.linspace(1,6,4):               # For loop to do several different values 
         
    data = oscillationmain(m,m,1,2,10)     # Obtaining the data given initial conditions
    
    file.write("\n\n" + "X values w/ intial conditions of A = 1, B = 2, X = " + str(m) + ", Y = " + str(m) + "\n\n")     
                                           # Creating a title for the loop in the file to distinguish initial conditions
                                           # of the list of data points
    
    for a in data[0]:                      # Loop writing all of the X values throughout the calculation to a file
        file.write(str(a) + " ")           # 
    
    file.write("\n\n" + "Y values w/ intial conditions of A = 1, B = 2, X = " + str(m) + ", Y = " + str(m) + "\n")          
        
    for b in data[1]:                      # Loop writing all of the Y values throughout the calculation to a file  
        file.write(str(b) + " ")           # 
        
    file.write("\n\n" + "T values corresponding to X and Y values above" + "\n")
    
    for c in data[2]:                      # Loop writing all of the T values throughout the calculation to a file
        file.write(str(b) + " ")           #
    
    py.xlabel("$x$")                                               #
    py.ylabel("$y$")                                               # 
    py.title("X vs Y for A,B = 1,2 and X,Y = %.2f,%.2f" %(m,m))    # Plotting
    py.plot(data[0], data[1],"r")                                  #
    py.show()                                                      #
    
# Opening and reading in the file to show that it contains the data from the program.
file = open("COdata.txt", "r")
print(file.read())

#---------------------------------------------------------------------------------------------------------------------------

#---------------------------------------------------------------------------------------------------------------------------
# Result
#--------
# From looking at the 4 different plots using increasing values of X and Y inbetween 1 and 10 we can see that as long as A 
# and B are the same, then we will obtain the same behavior independent of the values of X and Y as they all reach the same 
# results over time.
#---------------------------------------------------------------------------------------------------------------------------
```

```python
#---------------------------------------------------------------------------------------------------------------------------
# Part 3: Finding stable and unstable states
#---------------------------------------------------------------------------------------------------------------------------
# Goal:
#------
# Let the initial value of (X,Y ) equal (A + 0.001,B/A) for several dierent values of A and B, i.e., choose initial values 
# close to (but not equal!) the steady state values. Classify which initial values result in steady state behavior (stable) 
# and which ones show periodic behavior (unstable). Find the relation between A and B that separates the two types of 
# behavior. Make two plots of Y versus X, one for a stable case and one for an unstable case
#---------------------------------------------------------------------------------------------------------------------------
# Solution:
#---------------------------------------------------------------------------------------------------------------------------

# A modified version of the first oscillation function to take in a A and B value only and set the value of X and Y off of
# what A and B are and then performs the same integration. This way we are able to keep X and Y near the stable state values
# and change the inputed values for A and B to analyze behavior.
import numpy as np
from matplotlib import pylab as py

def oscillation(A,B):
    
    X = A + 0.001
    Y = B/A
    
    def f(r,t):
        X = r[0]
        Y = r[1]
        fx = A - (B+1)*X + X*X*Y
        fy = B*X - X*X*Y
        return np.array([fx,fy], float)

    # Array from start time 0 to end time of 30 with 1000 steps in between
    a = 0
    b = 100
    N = 10000
    h = b/N

    tpoints = np.arange(a,b,h)
    xpoints = []
    ypoints = []
    r = np.array([X,Y], float)
    
    # 4th-Order Runge-Kutta
    for t in tpoints:
        xpoints.append(r[0])
        ypoints.append(r[1])
        k1 = h*f(r,t)
        k2 = h*f(r+0.5*k1, t+0.5*h)
        k3 = h*f(r+0.5*k2, t+0.5*h)
        k4 = h*f(r+k3, t+h)
        r += (k1+ 2*k2 + 2*k3 + k4)/6
    
    return np.array([xpoints, ypoints, tpoints], float)
#---------------------------------------------------------------------------------------------------------------------------
#---------------------------------------------------------------------------------------------------------------------------
# Unstable Case Example:
#---------------------------------------------------------------------------------------------------------------------------
# Plot showing the an unstable scenario for the chemical oscillations. Showing periodic behavior
m = 1
data = oscillation(m,m*m+m)
py.xlabel("$x$")
py.ylabel("$y$")
py.title("X vs. Y for A,B = %.0f,%.0f" %(m,m*m+m))
py.plot(data[0],data[1],"g", label = 'Unstable')
py.legend()
py.show()

# Unstable because it never collapses down to a point therefore it is periodic oscillations
#---------------------------------------------------------------------------------------------------------------------------
```

```python
#---------------------------------------------------------------------------------------------------------------------------
# Stable Case Example:
#---------------------------------------------------------------------------------------------------------------------------
# Plot showing stable behavior in the chemical oscillation. No periodic oscillations
m = 3
data = oscillation(m,m*m+m)
py.xlabel("$x$")
py.ylabel("$y$")
py.title("X vs. Y for A,B = %.0f,%.0f" %(m,m*m+m))
py.plot(data[0],data[1],"m", label = 'Stable')
py.legend()
py.show()

# Stable because it eventually goes down to a point
#---------------------------------------------------------------------------------------------------------------------------
```

```python
#---------------------------------------------------------------------------------------------------------------------------
# Goal: Find Instability by fixing A and changing B
#---------------------------------------------------------------------------------------------------------------------------
# Solution:
#----------

# For loop to run though different iterations of X and Y and plot the results
# In this loop, A is a constant 1, but I am changing the values of  B to try to find instability
for m in np.linspace(1,3,3):
    data = oscillation(1,m)
    py.xlabel("$x$")
    py.ylabel("$y$")
    py.title("X vs. Y for A,B = 1," + str(int(m)))
    py.plot(data[0], data[1],"c")
    py.show()
#---------------------------------------------------------------------------------------------------------------------------
# Results:
#---------
# From analyzing the graphs we find that when A = 1 and B = 2 it results in unstable behavior. Otherwise all other cases are
# stable
#---------------------------------------------------------------------------------------------------------------------------
```
