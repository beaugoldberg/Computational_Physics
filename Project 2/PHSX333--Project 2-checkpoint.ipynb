```python
#--------------------------------------------------------------------------------------------------------------------------#
#-------------------------------------------------- Jacobi Method ---------------------------------------------------------#
#--------------------------------------------------------------------------------------------------------------------------#
import time
import numpy as np 
from matplotlib import pylab as py

def jacobi(M):
    start = time.time()          # Initial time stamp
    iterations = 0               # Variable to keep track of iterations
    accuracy = 1e-4              # Target Accuracy

    phi = np.zeros([M+1,M+1],float)          # Main array
    phiprime = np.zeros([M+1,M+1],float)     # Swap array
    
    #-------------------------------------------- Populating Main Array ---------------------------------------------------#
    for i in range(M+1):
        for j in range(M+1):
            if i == int((1/3)*M)//1 and j > int((1/4)*M)//1 and j < int((3/4)*M)//1:
                phi[i,j] = -1
                
            elif i == int((2/3)*M)//1 and j > int((1/4)*M)//1 and j < int((3/4)*M)//1:
                phi[i,j] = 1
                
    #----------------------------------------------------------------------------------------------------------------------#
    #----------------------------------------- Performing basic Jacobi Method ---------------------------------------------#
    #----------------------------------------------------------------------------------------------------------------------#
    delta = 1e-6+1
    while delta > accuracy:
       
        for i in range(M):
            for j in range(M):
                
                if i == 0 or i == M or j == 0 or j == M:
                    phiprime[i,j] = phi[i,j]
                    
                elif i == int((1/3)*M)//1 and j > int((1/4)*M)//1 and j < int((3/4)*M)//1:
                    phiprime[i,j] = -1
                    
                elif i == int((2/3)*M)//1 and j > int((1/4)*M)//1 and j < int((3/4)*M)//1:
                    phiprime[i,j] = 1
                
                else: phiprime[i,j] = (phi[i+1,j] + phi[i-1,j] + phi[i,j+1] + phi[i,j-1])/4

        delta = np.max(abs(phi-phiprime))
        phi,phiprime = phiprime,phi
        iterations += 1
        
    elapsed = time.time() - start
    return np.array([elapsed,iterations])
#--------------------------------------------------------------------------------------------------------------------------#
#--------------------------------------------------------------------------------------------------------------------------#
```

```python
#--------------------------------------------------------------------------------------------------------------------------#
#----------------------------------------------------- SOR Method ---------------------------------------------------------#
#--------------------------------------------------------------------------------------------------------------------------#
import time
import numpy as np
from matplotlib import pylab as py

def over_relax(Omega,M,d,option):
         
    start = time.time()                # Initial time stamp
    delta = 1e-6+1                     # Setting delta higher initially to enter while loop
    iterations = 0                     # Variable to keep track of iterations
    accuracy = 1e-4
    
    phi = np.zeros([M+1,M+1],float)    # Main array
    
    #--------------------------------------------- Populating Main Array --------------------------------------------------#
    for i in range(1,M):
        for j in range(1,M):
                
            if i == int((d*M))//1 and j > int((1/4)*M)//1 and j < int((3/4)*M)//1:
                phi[i,j] = -1

            elif i == M - int((d*M))//1 and j > int((1/4)*M)//1 and j < int((3/4)*M)//1:
                phi[i,j] = 1
                
    #----------------------------------------------------------------------------------------------------------------------#
    #---------------------------------------------- Performing SOR Method -------------------------------------------------#
    #----------------------------------------------------------------------------------------------------------------------#
    while delta > accuracy:
    
        delta = 0
        for i in range(1,M):
            for j in range(1,M):
                
                if i == int((d*M))//1 and j > int((1/4)*M)//1 and j < int((3/4)*M)//1:
                    phi[i,j] = -1

                elif i == M - int((d*M))//1 and j > int((1/4)*M)//1 and j < int((3/4)*M)//1:
                    phi[i,j] = 1
                
                else:
                    potentialdifference = (phi[i+1,j] + phi[i-1,j] + phi[i,j+1] + phi[i,j-1])/4 - phi[i,j]
                    phi[i,j] = phi[i,j] + (Omega)*potentialdifference 
                
                if potentialdifference > delta:
                    delta = potentialdifference
                 
        iterations += 1
    
    elapsed = time.time() - start
    if option == 1: return np.array([elapsed, Omega, iterations])
    if option == 2: return phi
#--------------------------------------------------------------------------------------------------------------------------#
#--------------------------------------------------------------------------------------------------------------------------#
```

```python
# -------------------------------------------------------------------------------------------------------------------------#
#-------------------------------------- SOR Iterations of Over-Relaxation Variables ---------------------------------------#
#--------------------------------------------------------------------------------------------------------------------------#
omegas = [1.0,1.05,1.1,1.15,1.2,1.25,1.3,1.35,1.4,1.45,1.5]
iterations = []
for om in omegas:
    data = over_relax(om,100,1/3,1)
    iterations.append(data[2])
    print("Time elapsed: %.2f seconds" %(data[0]))
    print("Over-relaxation variable: %.2f" %(data[1]))
    print("Iterations until convergence: " + str(data[2]) + "\n")
    
py.plot(omegas,iterations,"o-")
py.title("Omegas vs. Iterations Until Convergence with 100x100 Grid")
py.ylabel("$Iterations$") ; py.xlabel("$Omegas(w)$")
py.show()
#--------------------------------------------------------------------------------------------------------------------------#
#--------------------------------------------------------------------------------------------------------------------------#
```

```python
#--------------------------------------------------------------------------------------------------------------------------#
#------------------------------------------- SOR Iterations of Grid Values ------------------------------------------------#
#--------------------------------------------------------------------------------------------------------------------------#
Grid_Spaces = np.array([50,100,200,400])                          # Array of Grid size values to iterate over
Iteration_Values = []                                             # Array to save iteration values used to plot
for grid in Grid_Spaces:
    data = over_relax(0.5,grid,1/3,1)
    Iteration_Values.append(data[2])
    print("Number of grid spaces: " + str(grid))
    print("Time elapsed: %.2f seconds" %(data[0]))
    print("Iterations until convergence: " + str(data[2]) + "\n")

py.plot(Grid_Spaces, Iteration_Values,"o-")
py.ylabel("$Iterations$") ; py.xlabel("$Grid Spaces$")
py.title("Number of Iterations vs. Grid Spaces using SOR")
py.show()
#--------------------------------------------------------------------------------------------------------------------------#
#--------------------------------------------------------------------------------------------------------------------------#
```

```python
def SOR_field(phi,M,d):
    
    Ex_Field = np.zeros([M,M],float)
    Ey_Field = np.zeros([M,M],float)
    h = 2/M
    for i in range(1,M-1):
        for j in range(1,M-1):
            Ex_Field[i,j] = -(phi[i+1,j] - phi[i-1,j])/h
            Ey_Field[i,j] = -(phi[i,j+1] - phi[i,j-1])/h
    
    skip = (slice(None,None,2),slice(None,None,2))
    X,Y = np.meshgrid(np.linspace(0,M),np.linspace(0,M))
    py.quiver(X[skip],Y[skip],Ex_Field[skip].T,Ey_Field[skip].T)
    py.title("Electric Field w/ Plates " + str(d) + "m from edge")
    py.show()

distances = [.1,.2,.3,.4]
for d in distances:
    phi = over_relax(0.5,50,d,2)
    SOR_field(phi,50,d)
```
