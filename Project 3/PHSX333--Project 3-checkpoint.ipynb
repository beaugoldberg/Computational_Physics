```python
#--------------------------------------------------------------------------------------------------------------------------#
#-------------------------------------------- Project 3: Self Avoiding Random Walk ----------------------------------------#
#--------------------------------------------------------------------------------------------------------------------------#
import numpy as np
import random
import math
import os
import matplotlib.pylab as py
from scipy.optimize import curve_fit
#--------------------------------------------------------------------------------------------------------------------------#
#-------------------------------------- Main Function: Does the Random Walk (Non-Weighted) --------------------------------#
#--------------------------------------------------------------------------------------------------------------------------#
def walker(walks,walkers,plot):
   
    py.rcParams['axes.facecolor'] = 'black'     #Changes plot background to black
    successful = 0                              #Keeps track of successful chains (doesnt run into itself)
    distances = 0                               #Used to calculate the average distance squared for the program
    M = 2*walks                                 #Grid Size
    files = []                                  #Used in making the animation
    ii = 0                                      #Used in making the animation
    
    for j in range(walkers): 
        N = walks                                                     #Steps in a walk
        x_values = [] ; y_values = []                                 #Hold x and y values for plotting
        lattice = np.zeros([M+1,M+1],float)                           #Set up grid for walk
        origin = int(M/2) ; x = int(M/2) ; y = int(M/2)               #Set initial origin, x, y
        lattice[x,y] = 1 ; x_values.append(x) ; y_values.append(y)    #Append initial position
        movement = 1 ; y += 1                                         #Take first step
        lattice[x,y] = 1 ; x_values.append(x) ; y_values.append(y)    #Append second position
           
        i = 1                                       #Begin Walk
        while i < N:                                #Movement Code == 0-Left, 1-Up, 2-Right, 3-Down
            num = random.random()*3                 
            if 0 <= num < 1:                        #Left Movement from Forward
                if movement == 0: 
                    y += -1 ; movement = 3 
                    if lattice[x,y] == 0: lattice[x,y] = 1 ; i += 1 ; x_values.append(x) ; y_values.append(y) 
                    else: y += 1 ; i = N+1
                elif movement == 1: 
                    x += -1 ; movement = 0
                    if lattice[x,y] == 0: lattice[x,y] = 1 ; i += 1 ; x_values.append(x) ; y_values.append(y) 
                    else: x += 1 ; i = N+1
                elif movement == 2: 
                    y += 1 ; movement = 1
                    if lattice[x,y] == 0: lattice[x,y] = 1 ; i += 1 ; x_values.append(x) ; y_values.append(y) 
                    else: y += -1 ; i = N+1
                elif movement == 3:
                    x += 1 ; movement = 2
                    if lattice[x,y] == 0: lattice[x,y] = 1 ; i += 1 ; x_values.append(x) ; y_values.append(y) 
                    else: x += -1 ; i = N+1
                        
            elif 1 <= num < 2:                      #Forward Movement from Forward
                if movement == 0: 
                    x += -1
                    if lattice[x,y] == 0: lattice[x,y] = 1 ; i += 1 ; x_values.append(x) ; y_values.append(y) 
                    else: x += 1 ; i = N+1
                elif movement == 1: 
                    y += 1
                    if lattice[x,y] == 0: lattice[x,y] = 1 ; i += 1 ; x_values.append(x) ; y_values.append(y) 
                    else: y += -1 ; i = N+1
                elif movement == 2: 
                    x += 1 
                    if lattice[x,y] == 0: lattice[x,y] = 1 ; i += 1 ; x_values.append(x) ; y_values.append(y) 
                    else: x += -1 ; i = N+1
                elif movement == 3: 
                    y += -1
                    if lattice[x,y] == 0: lattice[x,y] = 1 ; i += 1 ; x_values.append(x) ; y_values.append(y) 
                    else: y += 1 ; i = N+1

            elif 2 <= num < 3:                      #Right Movement from Forward
                if movement == 0:                                              
                    y += 1 ; movement = 1
                    if lattice[x,y] == 0: lattice[x,y] = 1 ; i += 1 ; x_values.append(x) ; y_values.append(y)     
                    else: y += -1 ; i = N+1                                                             
                elif movement == 1:
                    x += 1 ; movement = 2
                    if lattice[x,y] == 0: lattice[x,y] = 1 ; i += 1 ; x_values.append(x) ; y_values.append(y) 
                    else: x += -1 ; i = N+1
                elif movement == 2: 
                    y += -1 ; movement = 3
                    if lattice[x,y] == 0: lattice[x,y] = 1 ; i += 1 ; x_values.append(x) ; y_values.append(y) 
                    else: y += 1 ; i = N+1
                elif movement == 3: 
                    x += -1 ; movement = 0 
                    if lattice[x,y] == 0: lattice[x,y] = 1 ; i += 1 ; x_values.append(x) ; y_values.append(y) 
                    else: x += 1 ; i = N+1
                        
        d = (abs(x-origin)**2 + abs(y-origin)**2)                     #Calculates distance squared and adds to variable
        if i == N: successful += 1  ; distances += d                  #If chain finished all steps it is marked successful
        if plot == 1:                                                 #Making the animation
            ii += 1
            py.plot(x_values,y_values)
            py.title("Self_Avoiding Random Walk showing %.0f Walkers of %.0f steps" %(walkers,walks))
            py.xlim(origin-walks,origin+walks) ; py.ylim(origin-walks,origin+walks)
            fname = '_tmp%03d.png' % ii
            py.savefig(fname) 
            files.append(fname)  
                   
    p_result = (successful/walkers)*100                                #Gives final percentage of successful walkers
    if successful == 0: d_result = 0
    else: d_result = distances/successful                              #Gives final average squared distance 
    
    if plot == 1:                                                      #More animation code
        os.system("ffmpeg -i _tmp%03d.png -c:v libx264 -r 30 -pix_fmt yuv420p unweighted30.mp4")
        for fname in files: os.remove(fname) 
        py.show()
    
    return p_result,d_result
#--------------------------------------------------------------------------------------------------------------------------#
#------------ Function for calculating average percentage and distance over 'N' amount of trials (For Accuracy) -----------#
#--------------------------------------------------------------------------------------------------------------------------#
def avg_walk(trials,walks,walkers,test,plot):
    percentages = 0
    avg_distance = 0
    for i in range(trials): percent,d = walker(walks,walkers,plot) ; percentages += percent ; avg_distance += d
    p_result = percentages/trials
    d_result = avg_distance/trials
    
    if test == 0: return trials,walks,walkers,p_result,d_result       #Option for printing results easier 
    if test == 1: return p_result,d_result                            #Option used for plotting results
#--------------------------------------------------------------------------------------------------------------------------#
#------------------------------ Simple test to print percentages and length of user chosen numbers ------------------------#
#--------------------------------------------------------------------------------------------------------------------------#
trials,walks,walkers,p_result,d_result = avg_walk(1,30,250,0,1)
print("The average success rate after %.0f trials of %.0f walkers with a length of %.0f monomers is %.2f percent." \
      % (trials,walkers,walks,p_result))
print("The mean distance sqaured from origin after %.0f trials of %.0f chains of %.0f monomers each is %.2f monomers." \
      %(trials,walkers,walks,d_result))
#--------------------------------------------------------------------------------------------------------------------------#
#--------------------------------------------------------------------------------------------------------------------------#
```

```python
#--------------------------------------------------------------------------------------------------------------------------#
#-------------------------------------------- Project 3: Self Avoiding Random Walk ----------------------------------------#
#--------------------------------------------------------------------------------------------------------------------------#
import numpy as np
import random
import math
import os
import matplotlib.pylab as py
from scipy.optimize import curve_fit
#--------------------------------------------------------------------------------------------------------------------------#
#---------------------------------------- Main Function: Does the Random Walk (Weighted) ----------------------------------#
#--------------------------------------------------------------------------------------------------------------------------#
def walker_w(walks,walkers,plot):
   
    py.rcParams['axes.facecolor'] = 'black'              #Makes plot background black
    successful = 0                                       #Keeps track of successful walks
    weights = []
    distances = []
    M = 2*walks                                          #Grid size
    files = []                                           #Used in making animation
    ii = 0                                               #Used in making animation
    
    for j in range(walkers): 
        W = 1                                                        #Initial weight
        N = walks                                                    #Steps in a walk
        x_values = [] ; y_values = []                                #Holds X and Y positions for plotting
        lattice = np.zeros([M+1,M+1],float)                          #Creates grid for the walk
        origin = int(M/2) ; x = int(M/2) ; y = int(M/2)              #Initializes origin, x, y
        lattice[x,y] = 1 ; x_values.append(x) ; y_values.append(y)   #Appends intitial position
        movement = 1 ; y += 1                                        #Takes first step
        lattice[x,y] = 1 ; x_values.append(x) ; y_values.append(y)   #Append second position
           
        i = 1                                     #Begin walk
        while i < N:                              #Movement Code == 0-Left, 1-Up, 2-Right, 3-Down
            num = random.random()*3
            if 0 <= num < 1:                      #Left Movement from Forward
                if movement == 0: 
                    y += -1 ; movement = 3 
                    if lattice[x,y] == 0: lattice[x,y] = 1 ; i += 1 ; x_values.append(x) ; y_values.append(y) 
                    else: y += 1 ; i,x,y,lattice,movement,x_values,y_values,W = weight(0,0,lattice,x,y,i,x_values,y_values,N,W)
                elif movement == 1: 
                    x += -1 ; movement = 0
                    if lattice[x,y] == 0: lattice[x,y] = 1 ; i += 1 ; x_values.append(x) ; y_values.append(y) 
                    else: x += 1 ; i,x,y,lattice,movement,x_values,y_values,W = weight(0,1,lattice,x,y,i,x_values,y_values,N,W)
                elif movement == 2: 
                    y += 1 ; movement = 1
                    if lattice[x,y] == 0: lattice[x,y] = 1 ; i += 1 ; x_values.append(x) ; y_values.append(y) 
                    else: y += -1; i,x,y,lattice,movement,x_values,y_values,W =  weight(0,2,lattice,x,y,i,x_values,y_values,N,W)
                elif movement == 3:
                    x += 1 ; movement = 2
                    if lattice[x,y] == 0: lattice[x,y] = 1 ; i += 1 ; x_values.append(x) ; y_values.append(y) 
                    else: x += -1 ; i,x,y,lattice,movement,x_values,y_values,W = weight(0,3,lattice,x,y,i,x_values,y_values,N,W)
                        
            elif 1 <= num < 2:                    #Forward Movement from Forward
                if movement == 0: 
                    x += -1
                    if lattice[x,y] == 0: lattice[x,y] = 1 ; i += 1 ; x_values.append(x) ; y_values.append(y) 
                    else: x += 1 ; i,x,y,lattice,movement,x_values,y_values,W = weight(1,0,lattice,x,y,i,x_values,y_values,N,W)
                elif movement == 1: 
                    y += 1
                    if lattice[x,y] == 0: lattice[x,y] = 1 ; i += 1 ; x_values.append(x) ; y_values.append(y) 
                    else: y += -1 ; i,x,y,lattice,movement,x_values,y_values,W = weight(1,1,lattice,x,y,i,x_values,y_values,N,W)
                elif movement == 2: 
                    x += 1 
                    if lattice[x,y] == 0: lattice[x,y] = 1 ; i += 1 ; x_values.append(x) ; y_values.append(y) 
                    else: x += -1 ; i,x,y,lattice,movement,x_values,y_values,W = weight(1,2,lattice,x,y,i,x_values,y_values,N,W)
                elif movement == 3: 
                    y += -1
                    if lattice[x,y] == 0: lattice[x,y] = 1 ; i += 1 ; x_values.append(x) ; y_values.append(y)
                    else: y += 1 ; i,x,y,lattice,movement,x_values,y_values,W = weight(1,3,lattice,x,y,i,x_values,y_values,N,W)

            elif 2 <= num < 3:                    #Right Movement from forward
                if movement == 0:                                              
                    y += 1 ; movement = 1
                    if lattice[x,y] == 0: lattice[x,y] = 1 ; i += 1 ; x_values.append(x) ; y_values.append(y)
                    else: y += -1 ; i,x,y,lattice,movement,x_values,y_values,W = weight(2,0,lattice,x,y,i,x_values,y_values,N,W)
                elif movement == 1:
                    x += 1 ; movement = 2
                    if lattice[x,y] == 0: lattice[x,y] = 1 ; i += 1 ; x_values.append(x) ; y_values.append(y)
                    else: x += -1 ; i,x,y,lattice,movement,x_values,y_values,W = weight(2,1,lattice,x,y,i,x_values,y_values,N,W)
                elif movement == 2: 
                    y += -1 ; movement = 3
                    if lattice[x,y] == 0: lattice[x,y] = 1 ; i += 1 ; x_values.append(x) ; y_values.append(y)
                    else: y += 1 ; i,x,y,lattice,movement,x_values,y_values,W = weight(2,2,lattice,x,y,i,x_values,y_values,N,W)
                elif movement == 3: 
                    x += -1 ; movement = 0 
                    if lattice[x,y] == 0: lattice[x,y] = 1 ; i += 1 ; x_values.append(x) ; y_values.append(y)
                    else: x += 1 ; i,x,y,lattice,movement,x_values,y_values,W = weight(2,3,lattice,x,y,i,x_values,y_values,N,W)
                  
        d = (abs(x-origin)**2 + abs(y-origin)**2) ; distances.append(d) ; weights.append(W)
        if i == N: successful += 1                                       #If chain finishes steps it is marked successful
        if plot == 1:                                                    #Making animation
            ii += 1
            py.plot(x_values,y_values)
            py.title("Self Avoiding Random Walk showing %.0f Walkers of %.0f steps" %(walkers,walks))
            py.xlim(origin-walks,origin+walks) ; py.ylim(origin-walks,origin+walks)
            fname = '_tmp%03d.png' % ii
            py.savefig(fname) 
            files.append(fname)  
                   
    p_result = (successful/walkers)*100                                   #Calulates percentage of successful walkers
    d_result = np.average(distances, weights = weights)                   #Mean distance squared
    
    if plot == 1:                                                         #More animation code
        os.system("ffmpeg -i _tmp%03d.png -c:v libx264 -r 30 -pix_fmt yuv420p weighted50.mp4")
        for fname in files: os.remove(fname) 
        py.show()
    
    return p_result,d_result
#--------------------------------------------------------------------------------------------------------------------------#
#--------------------------- Weight Function: Determines the Step Options and Takes Step if Possible ----------------------#
#--------------------------------------------------------------------------------------------------------------------------#
def weight(choice,movement,a,x,y,i,x_values,y_values,N,W):
    #Choice code -- 0 = Left , 1 = Forward , 2 = Right
    #Movement code -- 0 = Left, 1 = Forward , 2 = Right, 3 = Down
    opt = []
    if choice == 0:
        if movement == 0:  #desired move is down but not an option 
            if a[x-1,y] == 0: opt.append(0)
            elif a[x,y+1] == 0: opt.append(1)
            end,W = option(opt,W)
            if end == 0: x += -1 ; movement = 0 ; a[x,y] = 1 ; i += 1 ; x_values.append(x) ; y_values.append(y)
            elif end == 1: y += 1 ; movement = 1 ; a[x,y] = 1 ; i += 1 ; x_values.append(x) ; y_values.append(y)
            else: i = N+1 ; return i,x,y,a,movement,x_values,y_values,W
        elif movement == 1: #desired move is left but not an option
            if a[x,y+1] == 0: opt.append(1)
            elif a[x+1,y] == 0: opt.append(2)
            end,W = option(opt,W)
            if end == 1: y += 1 ; movement = 1 ; a[x,y] = 1 ; i += 1 ; x_values.append(x) ; y_values.append(y)
            elif end == 2: x += 1 ; movement = 2 ; a[x,y] = 1 ; i += 1 ; x_values.append(x) ; y_values.append(y)
            else: i = N+1 ; return i,x,y,a,movement,x_values,y_values,W
        elif movement == 2:#desired move is up but not an option
            if a[x+1,y] == 0: opt.append(2)
            elif a[x,y-1] == 0: opt.append(3)
            end,W = option(opt,W)
            if end == 2: x += 1 ; movement = 2 ; a[x,y] = 1 ; i += 1 ; x_values.append(x) ; y_values.append(y)
            elif end == 3: y += -1 ; movement = 3 ; a[x,y] = 1 ; i += 1 ; x_values.append(x) ; y_values.append(y)
            else: i = N+1 ; return i,x,y,a,movement,x_values,y_values,W
        elif movement == 3:#desired move is right but not an option
            if a[x,y-1] == 0: opt.append(3)
            elif a[x-1,y] == 0: opt.append(0)
            end,W = option(opt,W)
            if end == 3: y += -1 ; movement = 3 ; a[x,y] = 1 ; i += 1 ; x_values.append(x) ; y_values.append(y)
            elif end == 0: x += -1 ; movement = 0 ; a[x,y] = 1 ; i += 1 ; x_values.append(x) ; y_values.append(y)
            else: i = N+1 ; return i,x,y,a,movement,x_values,y_values,W
    elif choice == 1:
        if movement == 0: #desired move is left but not an option
            if a[x,y+1] == 0: opt.append(1)
            elif a[x,y-1] == 0: opt.append(3)
            end,W = option(opt,W)
            if end == 1: y += 1 ; movement = 1 ; a[x,y] = 1 ; i += 1 ; x_values.append(x) ; y_values.append(y)
            elif end == 3: y += -1 ; movement = 3 ; a[x,y] = 1 ; i += 1 ; x_values.append(x) ; y_values.append(y)
            else: i = N+1 ; return i,x,y,a,movement,x_values,y_values,W
        elif movement == 1:#desired move is up but not an option
            if a[x-1,y] == 0: opt.append(0)
            elif a[x+1,y] == 0: opt.append(2)
            end,W = option(opt,W)
            if end == 0: x += -1 ; movement = 0 ; a[x,y] = 1 ; i += 1 ; x_values.append(x) ; y_values.append(y)
            elif end == 2: x += 1 ; movement = 2 ; a[x,y] = 1 ; i += 1 ; x_values.append(x) ; y_values.append(y)
            else: i = N+1 ; return i,x,y,a,movement,x_values,y_values,W
        elif movement == 2:#desired move is right but not an option
            if a[x,y+1] == 0: opt.append(1)
            elif a[x,y-1] == 0: opt.append(3)
            end,W = option(opt,W)
            if end == 1: y += 1 ; movement = 1 ; a[x,y] = 1 ; i += 1 ; x_values.append(x) ; y_values.append(y)
            elif end == 3: y += -1 ; movement = 3 ; a[x,y] = 1 ; i += 1 ; x_values.append(x) ; y_values.append(y)
            else: i = N+1 ; return i,x,y,a,movement,x_values,y_values,W
        elif movement == 3:#desired move is down but not an option
            if a[x-1,y] == 0: opt.append(0)
            elif a[x+1,y] == 0: opt.append(2)
            end,W = option(opt,W)
            if end == 0: x += -1 ; movement = 0 ; a[x,y] = 1 ; i += 1 ; x_values.append(x) ; y_values.append(y)
            elif end == 2: x += 1 ; movement = 2 ; a[x,y] = 1 ; i += 1 ; x_values.append(x) ; y_values.append(y)
            else: i = N+1 ; return i,x,y,a,movement,x_values,y_values,W
    elif choice == 2:
        if movement == 0:#desired move is up but not an option
            if a[x-1,y] == 0: opt.append(0)
            elif a[x,y-1] == 0: opt.append(3)
            end,W = option(opt,W)
            if end == 0: x += -1 ; movement = 0 ; a[x,y] = 1 ; i += 1 ; x_values.append(x) ; y_values.append(y)
            elif end == 3: y += -1 ; movement = 3 ; a[x,y] = 1 ; i += 1 ; x_values.append(x) ; y_values.append(y)
            else: i = N+1 ; return i,x,y,a,movement,x_values,y_values,W
        elif movement == 1:#desired move is right but not an option
            if a[x,y+1] == 0: opt.append(1)
            elif a[x-1,y] == 0: opt.append(0)
            end,W = option(opt,W)
            if end == 1: y += 1 ; movement = 1 ; a[x,y] = 1 ; i += 1 ; x_values.append(x) ; y_values.append(y)
            elif end == 0: x += -1 ; movement = 0 ; a[x,y] = 1 ; i += 1 ; x_values.append(x) ; y_values.append(y)
            else: i = N+1 ; return i,x,y,a,movement,x_values,y_values,W
        elif movement == 2:#desired move is down but not an option
            if a[x+1,y] == 0: opt.append(2)
            elif a[x,y+1] == 0: opt.append(1)
            end,W = option(opt,W)
            if end == 2: x += 1 ; movement = 2 ; a[x,y] = 1 ; i += 1 ; x_values.append(x) ; y_values.append(y)
            elif end == 1: y += 1 ; movement = 1 ; a[x,y] = 1 ; i += 1 ; x_values.append(x) ; y_values.append(y)
            else: i = N+1 ; return i,x,y,a,movement,x_values,y_values,W
        elif movement == 3:#desired move is left but not an option
            if a[x,y-1] == 0: opt.append(3)
            elif a[x+1,y] == 0: opt.append(2)
            end,W = option(opt,W)
            if end == 3: y += -1 ; movement = 3 ; a[x,y] = 1 ; i += 1 ; x_values.append(x) ; y_values.append(y)
            elif end == 2: x += 1 ; movement = 2 ; a[x,y] = 1 ; i += 1 ; x_values.append(x) ; y_values.append(y)
            else: i = N+1 ; return i,x,y,a,movement,x_values,y_values,W
    return i,x,y,a,movement,x_values,y_values,W
#--------------------------------------------------------------------------------------------------------------------------#
#----------------------------- Option function: decides which step to take based on possibilities -------------------------#
#--------------------------------------------------------------------------------------------------------------------------#
def option(opt,W):
    if len(opt) == 0: end_choice = 12 ; W = 0
    elif len(opt) == 1: end_choice = opt[0] ; W = (1/3)*W
    elif len(opt) == 2:
        W = (2/3)*W
        num = random.random*2
        if 0 <= num < 1: end_choice = opt[0]
        else: end_choice = opt[1]
    return end_choice,W
#--------------------------------------------------------------------------------------------------------------------------#
#------------ Avg_Walk Function: Determines Mean Distance Squared and Percentage for 'N' trials (For Accuracy) ------------#
#--------------------------------------------------------------------------------------------------------------------------#
def avg_walk_w(trials,walks,walkers,test,plot):
    avg_percentages = 0
    avg_distance = 0
    for i in range(trials): percent,d = walker_w(walks,walkers,plot) ; avg_percentages += percent ; avg_distance += d
    p_result = avg_percentages/trials
    d_result = avg_distance/trials
    
    if test == 0: return trials,walks,walkers,p_result,d_result       #Option for printing results 
    if test == 1: return p_result,d_result                            #Option used for plotting results
#--------------------------------------------------------------------------------------------------------------------------#
#------------------------------------------------------ Test --------------------------------------------------------------#
#--------------------------------------------------------------------------------------------------------------------------#
trials,walks,walkers,p_result,d_result = avg_walk_w(1,50,250,0,1)
print("The average success rate after %.0f trials of %.0f walkers with a length of %.0f monomers is %.2f percent." \
      % (trials,walkers,walks,p_result))
print("The mean distance squared from origin after %.0f trials of %.0f chains of %.0f monomers each is %.2f monomers." \
      %(trials,walkers,walks,d_result))
#--------------------------------------------------------------------------------------------------------------------------#
#--------------------------------------------------------------------------------------------------------------------------#
```
