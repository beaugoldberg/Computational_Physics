```python
#---------------------------------------------------------------------------------------------------------------------------
# Exercise 9.4: Thermal Diffusion at the Earths Crust (FTCS Method)
#---------------------------------------------------------------------------------------------------------------------------
import numpy as np 
from matplotlib import pylab as py

tau = 365     # Days
T0 = 10       # Initial temperature in celsius
TMid = 11     # Avergae temperature in celsius
TMax = 12     # Max temperature in celsius
D = 0.1       # Diffusion constant (m^2/day)
L = 20        # Thickness of material in meters
N = 100       # Number of divisions in the grid
a = L/N       # Grid Spacing
t0 = 0        # Initial time in seconds
tMax = tau*9  # Max time in seconds
h = 0.01      # Time step in seconds
c = h*D/a**2  # Iteration Constant

def Temp(T):                                  # Calculating initial temperature values
    return T0 + TMax*T*np.sin(2*np.pi*T/tau)

def f(T,t0,tMax):                             # Calculating our new temperature values
    while t0 < tMax:
        T[0] = Temp(t0)
        T[N] = 11
        T[1:N] = T[1:N] + c*(T[2:N+1]+T[0:N-1]-2*T[1:N])
        t0 += h   
    return T

T = np.empty(N+1, float)      # Initializing an array
Temp9Years = f(T,t0,tMax)     # Performing integration of our values over a nine year time period
t0 = tMax                     # Setting out tMax to 9 years to make next calculations more simple

P = np.zeros(4, float)        # Creating new array for iteration
for i in range(4):            # For loop to populate previous array with nine years plus three, six, then nine months
    P[i] = tMax+i*tau//4      #

for tMax in P:                                          #  
    Temp9Years = f(Temp9Years,t0,tMax)                  # Calculating temperature values at each of the
    py.plot(Temp9Years, label = '%.0f days' %(tMax))    # values in the P array
    t0 = tMax                                           #

py.xlabel("$x$")                                        #
py.ylabel("$T$")                                        #
py.title("Thermal Diffusion at the Earths Crust")       # Plotting
py.legend()                                             #
py.show()                                               #
#---------------------------------------------------------------------------------------------------------------------------
```

```python
#---------------------------------------------------------------------------------------------------------------------------
# Problem 2: The One Dimensional Wave Equation (Leapfrog Method)
#---------------------------------------------------------------------------------------------------------------------------
# We need to write down the leapfrog approximation for the both the right-hand side and the left-hand side
#
#      LHS: (d**2*u/d*t**2) = [u(x,t+a)+u(x,t-a)-2*u(x,t)]/a**2
#      RHS: (d**2*x/d*x**2) = [u(x+a,t)+u(x-a,t)-2*u(x,t)]/a**2
#
# We want to solve for u(x,t+a)
#
#      --> In our case c=1
#      ----> [u(x,t+dt)+u(x,t-dt)-2*u(x,t)]/dt**2 = (1**2)*[u(x+dx,t)+u(x-dx,t)-2*u(x,t)]/dx**2
#      -------> [u(x,t+dt)+u(x,t-dt)-2*u(x,t)] = (dt**2/dx**2)*[u(x+dx,t)+u(x-dx,t)-2*u(x,t)]
#      ---------> u(x,t+dt) = (dt**2/dx**2)*[u(x+dx,t)+u(x-dx,t)-2*u(x,t)]-u(x,t-dt)+2*u(x,t)
#
# We need to use a dt smaller than dx or the system will be unstable
#---------------------------------------------------------------------------------------------------------------------------
import numpy as np
from matplotlib import pylab as py
import copy

N = 100             # Number of grid points
L = 2               # Length of the material in meters
dx = .002           # Step size
dt = .001           # Step size
c = (dt/dx)**2      # Integration constant
epsilon = dt/1000   # Used for plotting

def u0(x):                                           # Function for the gaussian pulse
    return 0.1*np.e**(-(x-0.5)**2/(0.1)**2)

u = np.zeros(N+1, float)                             # Initializing arrays
v = np.zeros(N+1, float)                             #

for i in range(N):                                   # Populating the mid values for the u array
    u[i] = u0(2*i/N)

u[0] = 0; u[-1] = 0;                                 # Setting boundary values
py.plot(u,label = 't = 0.0s')                        # Plot initial results
u1 = copy.copy(u)                                    # Copying u array to use in the integration

v[1:-1] += (dt/dx**2)*(u[2:] - 2*u[1:-1] + u[0:-2])  # First step user the Euler-Cromer method
u[1:-1] += dt*v[1:-1]

# Times for plotting
t1 = 0; t2 = 0.5; t3 = 1.0; t4 = 1.5; t5 = 2.0; t6 = 2.5; t7 = 3.0; t8 = 3.5; t9 = 4.0; tend = t9 + epsilon

# Integration Loop
t = 0.0
while t < tend:
    for i in range(1,N):
        u1[i] = c*(u[i-1]-2*u[i]+u[i+1])-u1[i]+2*u[i]
    u,u1 = u1,u
    t += dt
    
    # Make plots at the given times
    if abs(t-t2)<epsilon:
        py.plot(u,label = 't = 0.5s')
    if abs(t-t3)<epsilon:
        py.plot(u,label = 't = 1.0s')
    if abs(t-t4)<epsilon:
        py.plot(u,label = 't = 1.5s')
    if abs(t-t5)<epsilon:
        py.plot(u,label = 't = 2.0s')
    if abs(t-t6)<epsilon:
        py.plot(u,label = 't = 2.5s')
    if abs(t-t7)<epsilon:
        py.plot(u,label = 't = 3.0s')
    if abs(t-t8)<epsilon:
        py.plot(u,label = 't = 3.5s')
    if abs(t-t9)<epsilon:
        py.plot(u,label = 't = 4.0s')

py.xlabel("$t$")
py.ylabel("$x$")
py.title("1-D Wave Equation: Fixed Ends")
py.legend()
py.show()
#---------------------------------------------------------------------------------------------------------------------------
```

```python
#---------------------------------------------------------------------------------------------------------------------------
# Problem 3: Burgers' Equation (FTCS Method)
#---------------------------------------------------------------------------------------------------------------------------
# We need to solve for u(x,t+dt)
#       
#       -------> du/dt = v*(d**2*u/d*x**2) - u*(du/dx)
#       -------> u(x,t+h) = u(x,t)+[v*dt/dx**2]*[u(x+dx,t)-2*u(x,t)+u(x-dx,t)]-u[i]*[u(x,t)-u(x-dx,t)]*dt/[2*dx]
#---------------------------------------------------------------------------------------------------------------------------
import numpy as np
from matplotlib import pylab as py
import copy

N = 200                # Number of grid points
L = 2                  # Length of string in meters
dx = .02               # Step size
dt = .001              # Step size
c = .05                # Viscosity
epsilon = dt/1000      # Used for plotting times
#--------------------------------------------------------------------------------------------------------------------------#
def u0(x,amp):                       # Funtion of our gaussian pulse
    return amp*np.e**(-(x-0.5)**2/(0.1)**2)

def f(a):         
    
    u = np.empty(N+1, float)         # Initializing an array
    for i in range(N):               # Populating the mid values for the u array
        u[i] = u0(2*i/N,a)

    u[0] = u[-2]; u[-1] = u[1];      # Setting boundary values
    py.plot(u,label = 't = 0.0s')    # Plot initial results
    u1 = np.empty(N+1,float)         # Initializing a second array

    # Times for plotting
    t1 = 0; t2 = 0.5; t3 = 1.0; t4 = 1.5; t5 = 2.0; t6 = 2.5; t7 = 3.0; t8 = 3.5; t9 = 4.0; tend = t9 + epsilon

    # Integration Loop
    t = 0.0
    while t < tend:
        for i in range(1,N):
            u1[i] = u[i]+(c*dt/dx**2)*(u[i+1]-2*u[i]+u[i-1])-u[i]*(u[i]-u[i-1])*dt/(2*dx)
            u[0] = u[-2]; u[-1] = u[1];
        u,u1 = u1,u
        t += dt

        # Make plots at the given times
        if abs(t-t2)<epsilon:
            py.plot(u,label = 't = 0.5s')
        if abs(t-t3)<epsilon:
            py.plot(u,label = 't = 1.0s')
        if abs(t-t4)<epsilon:
            py.plot(u,label = 't = 1.5s')
        if abs(t-t5)<epsilon:
            py.plot(u,label = 't = 2.0s')
        if abs(t-t6)<epsilon:
            py.plot(u,label = 't = 2.5s')
        if abs(t-t7)<epsilon:
            py.plot(u,label = 't = 3.0s')
        if abs(t-t8)<epsilon:
            py.plot(u,label = 't = 3.5s')
        if abs(t-t9)<epsilon:
            py.plot(u,label = 't = 4.0s')
    
    py.xlabel("$t$")
    py.ylabel("$x$")
    title = "Burgers' Equation: Amp = %.2f" %(a)
    py.title(title)
    py.legend()
    py.show()

P = [.1,2,6,10]     # Iterating though different amplitudes to analyze its effect on the plot
for i in P:         #
    f(i)            #
#---------------------------------------------------------------------------------------------------------------------------
# From analyzing the plots of Burgers' Equation using different amplitudes for the gaussian pulse, it seems that if the
# amplitude of the pulse is increased than the string will have more energy. Therefore it will take longer for the liquid 
# to dissapate the energy from the string, shown in the increase in time the amplitude of oscillations takes to reach 0.
#---------------------------------------------------------------------------------------------------------------------------
```
