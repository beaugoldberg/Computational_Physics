```python
#---------------------------------------------------------------------------------------------------------------------------
# Exercise 8.14: Quantum Oscillators
#---------------------------------------------------------------------------------------------------------------------------
#---------------------------------------------------------------------------------------------------------------------------
# Part A
#--------
# Goal: Write down the schrodinger equation for this problem as two first order equations. Write a program to find the 
# ground state energy and the energies of the first two excited states. 
#---------------------------------------------------------------------------------------------------------------------------
#
# If we use p = du/dx then we can get the following coupled first-order equations
#
#                             du/dx = p      ,      dp/dx = {(2*m*V0/hbar^2)(x^2/a^2)-(2*m*E/hbar^2)}*u      
#
#---------------------------------------------------------------------------------------------------------------------------
import numpy as np

# Constants
m = 9.1094e-31           # Mass of an electron
hbar = 1.0546e-34        # Plancks constant over 2 pi
e = 1.6022e-19           # Electron Charge
a = 1e-11   
N = 1000             
Vo = 50*e            
h = (10*a+10*a)/N    

def QO(x):
    
    if x == 0: 
        def V(x): return Vo*(x**2/a**2)
        
    if x == 1:
        def V(x): return Vo*(x**4/a**4)
        
    def f(r,x,E):
        p = r[0]
        u = r[1]
        fp = u
        fu = ((2*m/hbar**2)*(V(x)-E))*p
        return np.array([fp,fu], float)

    # Calculate the wavefunction for a particular energy
    def solve(E):
        p = 0.0
        u = 1.0
        r = np.array([p,u], float)

        # 4th-Order Runge-Kutta
        for x in np.arange(-10*a,10*a,h):
            k1 = h*f(r,x,E)
            k2 = h*f(r+0.5*k1,x+0.5*h,E)
            k3 = h*f(r+0.5*k2,x+0.5*h,E)
            k4 = h*f(r+k3,x+h,E)
            r += (k1+2*k2+2*k3+k4)/6

        return r[0]

    # Main program to find the energy using the secant method
    E1 = 0
    E2 = e
    E_States = []
    
    # Uses a specific method to find the energy states in the harmonic oscillator becuase it 
    # said energy levels are evenly spaced
    if x == 0:
        for i in range(3):
            p2 = solve(E1)

            target = e/N
            while abs(E1-E2) > target:
                p1,p2 = p2,solve(E2)
                E1,E2 = E2, E2-p2*(E2-E1)/(p2-p1) 

            E_States.append(E2/e)
            E1 = E2 + E_States[0] * e
            E2 = E1 + e
            
    # Uses a slightly different method then the previous to find energy states for the anharmonic oscillator 
    # becuase the energy states are no longer equally spaced
    if x == 1:
         for i in range(3):
            p2 = solve(E1)

            target = e/N
            while abs(E1-E2) > target:
                p1,p2 = p2,solve(E2)
                E1,E2 = E2, E2-p2*(E2-E1)/(p2-p1) 

            E_States.append(E2/e)
            E1 = E2 + E_States[i] * e
            E2 = E1 + e
        
    return E_States

E_States = QO(0)
print("Part A Values: " + "\n" + "----------------------")
print("Ground State Energy: %.3f eV" %(E_States[0]))
print("1st Excited State Energy: %.3f eV" %(E_States[1]))
print("2nd Excited State Energy: %.3f eV" %(E_States[2]))
print(" ")
print("Part B Values: "  + "\n" + "----------------------")
E_States = QO(1)
print("Ground State Energy: %.3f eV" %(E_States[0]))
print("1st Excited State Energy: %.3f eV" %(E_States[1]))
print("2nd Excited State Energy: %.3f eV" %(E_States[2]))
#---------------------------------------------------------------------------------------------------------------------------
```

```python
#---------------------------------------------------------------------------------------------------------------------------
# Exercise 9.1:
#---------------
# Write a program to solve Poisson's equation. Work in units of epsilon = 1 and continue the iteration until the solution 
# for the electric potential changes by less than 1e-6V per step at every grid point
#
#                                        Poissen's Equation 
#                                   ------------------------------
#                                       gradient^2(phi) = -rho
#
# Assuming epsilon = 1 in empty space as the problem suggests
#---------------------------------------------------------------------------------------------------------------------------

import numpy as np
from matplotlib import pylab as py

#Constants
M = 100             # Grid squares on a side
target = 10e-6      # Target accuracy
epsilon = 1         # Permittivity of free space 
a = 0.01

# Create arrays to hold potential values
phi = np.zeros([M+1,M+1],float)
phiprime = np.zeros([M+1,M+1],float)

def rho(x,y):
    if x > 0.6 and x < 0.8 and y > 0.6 and y < 0.8:          # The charge has to be 20 cm in from the edge and 20 cm wide 
        return 1                                             # in both x and y to be returned as a positive charge
    
    elif x > 0.2 and x < 0.4 and y > 0.2 and y < 0.4:        # The charge has a nuetral 20cm gap from the positive in the
        return -1                                            # 40cm to 60cm range and after that range is a 20cm gap that
                                                             # will return a negative charge
    
    else: return 0                                           # All other gaps (accounting for next to walls and inbetween 
                                                             # positive and negative charges) are nuetral    
# Main loop
delta = 1.0
while delta>target:

    # Calculate new values of the potential
    for i in range(1,M):
        for j in range(1,M):
            phiprime[i,j] = (phi[i+1,j] + phi[i-1,j] + phi[i,j+1] + phi[i,j-1])/4 + (a**2/(4*epsilon))*rho(i*a,j*a)

    # Calculate maximum difference from old values
    delta = np.max(abs(phi-phiprime))

    # Swap the two arrays around
    phi,phiprime = phiprime,phi

# Make a plot
py.imshow(phi, origin = 'lower')
py.gray()
py.show()
#---------------------------------------------------------------------------------------------------------------------------
```

```python
#---------------------------------------------------------------------------------------------------------------------------
# Exercise 9.3:
#---------------
# Write a program to calculate the electric potential in the box on a grid of 100 x 100 points, where the walls are at 
# voltage 0 and the two plates are at voltages +-1 as shown in the diagram. Use precision of 1e-6V and then make a density
# plot.
#---------------------------------------------------------------------------------------------------------------------------
import numpy as np
from matplotlib import pylab as py

#Constants
M = 100             # Grid squares on a side
V = 1.0             # Voltage at top of the wall
target = 1e-6       # Target accuracy

# Create arrays to hold potential values
phi = np.zeros([M+1,M+1],float)
   
# Main loop
delta = 1e-6+1      # Need to set an initial delta higher than our target to enter the following loop 
omega = 0.9         # Step value

while delta > target:

    delta = 0
    # Calculate new values of the potential
    for i in range(1,M):
        for j in range(1,M):
            if j == 20 and i > 20 and i < 80:
                phi[i,j] = 1
            
            elif j == 80 and i > 20 and i < 80:
                phi[i,j] = -1
            
            else:
                # The potetial difference is the difference between the potential of a gird point and its surrounding
                # neighbors. This is used to narrow down our accuracy of calculation because we want the potential
                # difference to be 1e-6
                potentialdifference = (phi[i+1,j] + phi[i-1,j] + phi[i,j+1] + phi[i,j-1])/4 - phi[i,j]
                phi[i,j] = phi[i,j] + (1+omega)*potentialdifference  # Taking a step
                
            if potentialdifference > delta :                # Stays in the loop until our target accuracy for delta 
                delta = potentialdifference                 # is reached 

# Make a plot
py.imshow(phi)
py.gray()
py.show()
#---------------------------------------------------------------------------------------------------------------------------
```
