```python
#PHSX 333:Computational Physics Homework 1 - Beau Goldberg
from math import pi

#Problem 1: Satellites and their orbits

#Newtons Law of Universal Gravitation is F = (G*M*m)/r**2 and the equation for centripital force is F = m*w**2*r which we can 
#convert using w = (2*pi)/T to F = (r*m*4*pi**2)/T**2 and we need to solve the two equations for r to get the height so 
#if we set the two equations equal we get (G*M*m)/r**2 = (r*m*4*pi**2)/T**2 and if you manipulate the equations to solve for r 
#you'll get r = ((G*m*T**2)/(4*pi**2))**1/3 and then you have to subtract the earths radius to get the altitude above the earth


print("Problem 1: Satellite Orbits")
print("---------------------------")
print("b.)")

T = float(input("Please enter the period of the satellite ... ")) #s
G = 6.67 * pow(10,-11)                                            #m^3kg^-1s^-2
M = 5.97 * pow(10,24)                                             #kg
R = 6371 * pow(10,3)                                              #m
height = pow((G*M*pow(T,2))/(4*pow(pi,2)), 1/3) - R               #m

OneDayPeriod = 24*60*60                   #s
NinetyMinPeriod = 90*60                   #s
FortyFiveMinPeriod = 45*60                #s
SiderealDayPeriod = 23*60*60 + 60*60*.98  #s

def altitude(T):
    altitude = pow((G*M*pow(T,2))/(4*pow(pi,2)), 1/3) - R 
    return altitude

print()
print("The altitude of a satellite orbiting in a",T,"second period is",height,"m")
print()
print("c.)")

OneDaySatellite = altitude(OneDayPeriod)               #One day is 86,400 seconds
NinetyMinSatellite = altitude(NinetyMinPeriod)         #90-min is 5400 seconds
FortyFiveMinSatellite = altitude(FortyFiveMinPeriod)   #45-min is 2700 seconds
SiderealDaySatellite = altitude(SiderealDayPeriod)     #Sidereal Day is 86,328 seconds
print("The altitude of a satellite orbiting in a one day period,",int(OneDayPeriod),"seconds, is",OneDaySatellite,"m")
print("The altitude of a satellite orbiting in a 90-minute period,",int(NinetyMinPeriod),"seconds, is", NinetyMinSatellite,"m")
print("The altitude of a satellite orbiting in a 45-minute period,",int(FortyFiveMinPeriod),"seconds, is", FortyFiveMinSatellite,"m")
print("The altitude of a satellite orbiting in a Sidereal Day period,",int(SiderealDayPeriod),"seconds, is",SiderealDaySatellite,"m")
#The result of the orbit for a 45 minute period ends up not being possible because the satellite would be inside of the earths radius

#The orbit is in a sidereal day because the earth completes one rotation in the 23.93 hours instead of 24 hours. It makes the altitude
#a little bit lower than a satellite in a 24 hour orbit
```

```python
#Problem 3: Binding Energy

print("Problem 3: Binding Energy")
print("-------------------------")
print("a.)")

a1 = 15.8     #MeV
a2 = 18.3     #MeV
a3 = 0.714    #MeV
a4 = 23.2     #MeV

def bindingenergy(A,Z):
    if A%2 == 0 and Z%2 == 0:
        a5 = 12
    if A%2 == 0 and Z%2 != 0:
        a5 = -12
    else:
        a5 = 0
        
    energy = a1*A - a2*pow(A,2/3) - a3*(pow(Z,2)/pow(A,1/3)) - a4*(pow(A-2*Z,2)/A) + a5/pow(A,0.5)
    return energy

A = int(input("Enter the mass number of the atom ... "))
Z = int(input("Enter the atomic number of the atom ... "))

#Binding energy per nucleon with given A and Z
print()
print("b.)")
print("The binding energy for a nucleus with mass number %i and atomic number %i is %f MeV" %(A,Z,bindingenergy(A,Z)))
print("The binding energy per nucleon is %f MeV" %(bindingenergy(A,Z)/A))

#Finding max A value to create max binding energy with a given Z
print()
print("c.)")

Z = int(input("Enter the atomic number of the atom ... "))
maxenergy = 0
maxA = 0
for a in range(Z,3*Z+1):
    energypernucleon = bindingenergy(a,Z)/a
    if energypernucleon > maxenergy:
        maxenergy = energypernucleon
        maxA = a
    
print("The value of A that results in the highest binding energy of %f MeV is %i" %(maxenergy, maxA)  )

#Max energy for random A and random Z
print()
print("d.)")
z=1
maxenergy = 0
maxA = 0

while z <=100:
    for a in range(z,3*z+1):
        energypernucleon = bindingenergy(a,z)/a
        if energypernucleon > maxenergy:
            maxenergy = energypernucleon
            maxA = a
        if energypernucleon == maxenergy:
            maxZ = z-1
            
    print("The max A value for Z = %i is %i resulting in a binding energy per nucleon of %f" %(z,maxA,maxenergy))
    z += 1
    
print()
print("The value for Z corresponding to the max binding energy is %i" %(maxZ))
```
